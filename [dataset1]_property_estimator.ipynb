###def function mape
- ค่าเกิน 9,10 ก็ลบ ศูนย์ ถ้าเกิน 2 คนประเมินว่าแบบนั้นเลย

```python
def mape(y_test, pred):
    y_test, pred = np.array(y_test), np.array(pred)
    mape = np.mean(np.abs((y_test - pred) / y_test))
    return mape
```

```python
#ใช้คอลัมน์ ประเมิน 1 ,ประเมิน 2 ในการคำนวณ
#คอลัมน์ Check1,2 ไว้เช็คว่าค่าพุ่งเกินไหม
table1= table['ประเมิน1']
table2= table['ประเมิน2']
table3= table['ราคาทรัพย์']
```

```python
mapetable = mape(table1, table2)
mapetable # ค่า mape ของผู้ประเมิน1 กับผู้ประเมิน 2
print(f'ความแตกต่างในการประเมินราคาทรัพย์แต่ละทรัพย์ของผู้ประเมินคนที่ 1 กับ ผู้ประเมิน 2 มีค่า MAPE อยู่ที่', mapetable)
```

```python
mapetable2 = mape(table1, table3)
mapetable2 # ค่า mape ของผู้ประเมิน1 กับ ราคาทรัพย์
print(f'ความแตกต่างในการประเมินราคาทรัพย์แต่ละทรัพย์ของผู้ประเมินคนที่ 1 กับ ราคาทรัพย์ มีค่า MAPE อยู่ที่', mapetable2)
```

```python
mapetable3 = mape(table2, table3)
mapetable3
print(f'ความแตกต่างในการประเมินราคาทรัพย์แต่ละทรัพย์ของผู้ประเมินคนที่ 2 กับ ราคาทรัพย์ มีค่า MAPE อยู่ที่', mapetable3)
```
